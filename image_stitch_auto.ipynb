import OpenCV, Num PY, Matplot

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Image Import and Read from left to right

```python
image1 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/CV project images/3.1.jpg')
image2 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/CV project images/3.2.jpg')
image3 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/CV project images/3.3.jpg')
image4 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/CV project images/3.4.jpg')
```

Image colour

```python
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)
image4 = cv2.cvtColor(image4, cv2.COLOR_BGR2RGB)
```

Figure size of Images

```python
plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
plt.imshow(image1)
plt.title('image1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image2)
plt.title('image2')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(image3)
plt.title('image3')
plt.axis('off')
plt.show()

plt.subplot(1, 4, 4)
plt.imshow(image4)
plt.title('image4')
plt.axis('off')
plt.show()
```

Panorama Image stitching tool

```python
stitcher = cv2.Stitcher_create()
```

check status for input images

```python
(status, stitched) = stitcher.stitch([image1, image2, image3, image4])
```

```python
if status == cv2.Stitcher_OK:
    print(" Stitching successful!")

    plt.figure(figsize=(15,8))
    plt.imshow(stitched)
    plt.title('Panorama Output')
    plt.axis('off')
    plt.show()

    # Save the stitched image
    cv2.imwrite('panorama_result.jpg', cv2.cvtColor(stitched, cv2.COLOR_RGB2BGR))

else:
    print(" Stitching failed with status:", status)
    # You can add more specific error handling based on status codes here if needed
    if status == cv2.Stitcher_ERR_NEED_MORE_IMGS:
        print("Error: Need more images to stitch.")
    elif status == cv2.Stitcher_ERR_NO_FEATURES:
        print("Error: Not enough features found in the images.")
    elif status == cv2.Stitcher_ERR_HOMOGRAPHY_EST_FAIL:
        print("Error: Homography estimation failed.")
    elif status == cv2.Stitcher_ERR_CAMERA_PARAMS_ADJUST_FAIL:
        print("Error: Camera parameters adjustment failed.")
```
